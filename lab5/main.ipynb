# Ex 1 - Preprocessing & Vectorizations

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess(doc):
    tokens = nltk.word_tokenize(doc.lower())
    tokens = [lemmatizer.lemmatize(t) for t in tokens if t.isalpha() and t not in stop_words]
    return ' '.join(tokens)

docs_clean = [preprocess(doc) for doc in docs]

# Bag-of-Words and TF-IDF
bagowords_vect = CountVectorizer()
tfidf_vect = TfidfVectorizer()

X_bow = bagowords_vect.fit_transform(docs_clean)
X_tfidf = tfidf_vect.fit_transform(docs_clean)

docs_clean
```

# Ex 2 - LSA SVD

```python
svd = TruncatedSVD(n_components=2, random_state=42)
X_bow_lsa = svd.fit_transform(X_bow)
X_tfidf_lsa = svd.fit_transform(X_tfidf)
```

# Ex 3 - NMF

```python
nmf = NMF(n_components=2, random_state=42)
X_bow_nmf = nmf.fit_transform(X_bow)
X_tfidf_nmf = nmf.fit_transform(X_tfidf)
```

# Ex 4 - LDA

```python
texts = [doc.split() for doc in docs_clean]
dictionary = Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]
lda = LdaModel(corpus=corpus, id2word=dictionary, num_topics=3, random_state=42)
```

# Ex 5 - Eval

```python
coherence_model_lda = CoherenceModel(model=lda, texts=texts, dictionary=dictionary, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()

print("LSA (BoW) shape:", X_bow_lsa.shape)
print("LSA (TF-IDF) shape:", X_tfidf_lsa.shape)
print("NMF (BoW) shape:", X_bow_nmf.shape)
print("NMF (TF-IDF) shape:", X_tfidf_nmf.shape)
print("LDA coherence score:", coherence_lda)

# 7. Scatter plots

output_dir = "images"
os.makedirs(output_dir, exist_ok=True)

def plot_scatter(X, title, labels, filename):
    plt.figure(figsize=(6,4))
    sns.scatterplot(x=X[:,0], y=X[:,1], hue=labels)
    plt.title(title)
    plt.savefig(os.path.join(output_dir, filename))
    plt.close()

plot_scatter(X_bow_lsa, "LSA (BoW)", topics, "lsa_bow.png")
plot_scatter(X_tfidf_lsa, "LSA (TF-IDF)", topics, "lsa_tfidf.png")
plot_scatter(X_bow_nmf, "NMF (BoW)", topics, "nmf_bow.png")
plot_scatter(X_tfidf_nmf, "NMF (TF-IDF)", topics, "nmf_tfidf.png")

# 8. Show LDA topics
for i, topic in lda.show_topics(num_topics=3, num_words=5, formatted=False):
    print(f"Topic {i}: {[word for word, prob in topic]}")
```
